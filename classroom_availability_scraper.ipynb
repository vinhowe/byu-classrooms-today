````python
def process_hours(a):
    days = []
#     print(a)
    for letter in a:
        days.append(False if letter == "_" else True)
#     print(days)
    return days

def group_hours_by_day(a):
#     day_hours = []
#     room_name = f"{a[0]} {a[1]}"
    for i in range(2, 17+1):
        a[i] = process_hours(a[i])
#     print(a)
    return a

# Return only the value for the hour of the specified day -- confusing but idc
def filter_hours_to_day(day_num):
    def input_day_applier(a):
        for i in range(2, 17+1):
            a[i] = process_hours(a[i])[day_num]
        return a
    return input_day_applier
````
